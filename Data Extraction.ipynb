```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

df = pd.read_excel('Test_Assignment/Input.xlsx')
if not os.path.exists('articles'):
    os.makedirs('articles')

for index, row in df.iterrows():
    url = row['URL']
    url_id = row['URL_ID']
    
    response = requests.get(url)
    html = response.content
    
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.find('title').text
    article_text = ''
    for p in soup.find_all('p'):
        article_text += p.text + '\n\n'
    
    article_text = article_text.replace('\n\n\n', '\n\n').strip()
    with open(f'articles/{url_id}.txt', 'w', encoding='utf-8') as f:
        f.write(title + '\n\n' + article_text)
    
    print(f'Extracted article {url_id} and saved to file.')
```
