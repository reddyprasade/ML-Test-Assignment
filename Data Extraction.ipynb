```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

df = pd.read_excel('Data/input.xlsx')
if not os.path.exists('articles'):
    os.makedirs('articles')

for index, row in df.iterrows():
    url = row['URL']
    url_id = row['URL_ID']
    
    response = requests.get(url)
    html = response.content
    
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.find('title').text
    article_text = ''
    for p in soup.find_all('p'):
        article_text += p.text + '\n\n'
    
    article_text = article_text.replace('\n\n\n', '\n\n').strip()
    with open(f'articles/{url_id}.txt', 'w', encoding='utf-8') as f:
        f.write(title + '\n\n' + article_text)
    
    print(f'Extracted article {url_id} and saved to file.')
```

```python
import os
import re

def load_stop_words(stop_words_folder):
    stop_words_file = os.path.join(stop_words_folder, 'stop_words.txt')
    with open(stop_words_file, 'r') as f:
        stop_words = set(f.read().splitlines())
    return stop_words

def load_master_dict(master_dict_folder):
    master_dict_file = os.path.join(master_dict_folder, 'master_dict.txt')
    with open(master_dict_file, 'r') as f:
        master_dict = {}
        for line in f.readlines():
            word, sentiment = line.strip().split('\t')
            master_dict[word] = sentiment
    return master_dict

def clean_text(text, stop_words):
    clean_text = ' '.join([word for word in text.split() if word not in stop_words])
    return clean_text

def create_sentiment_dicts(master_dict, stop_words):
    positive_words = [word for word, sentiment in master_dict.items() if sentiment == 'positive' and word not in stop_words]
    negative_words = [word for word, sentiment in master_dict.items() if sentiment == 'negative' and word not in stop_words]
    return positive_words, negative_words

def process_file(file_path, stop_words, master_dict):
    with open(file_path, 'r') as f:
        text = f.read()
    clean_text = clean_text(text, stop_words)
    positive_words, negative_words = create_sentiment_dicts(master_dict, stop_words)
    return clean_text, positive_words, negative_words


stop_words_folder = 'StopWords'
master_dict_folder = 'MasterDictionary'
stop_words = load_stop_words(stop_words_folder)
master_dict = load_master_dict(master_dict_folder)

files_folder = 'TextFiles'
files = [os.path.join(files_folder, f) for f in os.listdir(files_folder) if f.endswith('.txt')]

for file_path in files:
    clean_text, positive_words, negative_words = process_file(file_path, stop_words, master_dict)
    print(f'File: {file_path}')
    print('Clean Text:')
    print(clean_text)
    print('\nPositive Words:')
    print(positive_words)
    print('\nNegative Words:')
    print(negative_words)
    print('---')
```
